```python
import pandas as pd
import numpy as np

# Como tratar datos duplicados y/o datos perdidos.
#   

#Traemos las DF
data = pd.read_csv("data.csv.csv", nrows= 1e6)
data.head(10)

# DATOS DUPLICADOS.
# De no ser datos muy especificos los datos duplicados se suelen eliminar.

# Nuevo DF.
dataDuplicate = data.append(data)

# Ordenarlos.
dataDuplicate = dataDuplicate.sample(frac = 1)
dataDuplicate.head()

# Eliminar los datos faltantes
dataClean = dataDuplicate.drop_duplicates()
dataClean.head(10000000)

# Comprobamos la cantidad de filas de el DF creado y DF sin las diplicaciones >> yo me entiendo jaja :) <<  
len(dataDuplicate) == len(dataClean) # ouput: false.

# Eliminacion de datos un poco mas especifico.
# Es decir eliminar datos en una coluna en particular.
dataDuplicate2 = data.append(data)
dataDuplicate2 = dataDuplicate.sample(frac = 1)

dataClean = dataDuplicate2.drop_duplicates(subset= "age")
dataClean.head()

# Trabajo con datos faltantes.
data.dropna()

# Una especie de filtro, para una determinada cantidad de columnas con datos faltantes
data.dropna(thresh=len(data.columns) - 1)

# Para filtrar por una columna en especial.
data.dropna(subset= ["age", "sex"])
```
