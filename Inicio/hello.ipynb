```python
import pandas as pd
import numpy as np

# Con read_csv permite la carga de los datos al modelo. Pasandole como parametro la direccion de la BD
# BD del Titanic >> Creo
data = pd.read_csv("../data.csv.csv")

# IMPORTANTE.
# Cuando queremos usar funciones de nuestro Data Frame (DF) colocamos perentecis: Ej: data.head()
# Cuando lo que queremos es acceder a valores que ya estan en nuetro DF, no hace falta colocar parentecis. Ej: data.columns

# Con head() nos mostrar las primeras linas de la BD, pasando como argumento la cantidad de linas que se quiere
# Tener en cuanta que si no le pasamos un numero por parametro, por defecto mostrara las 5 primeras filas
data.head(3)

# Con la funcion .tail(), hacemos lo mismo quen con head() pero opteniendo los n filas inferiores, De forma ordenada
data.tail(5)

# Con el metodo sample() ordenara los datos de forma aleatoria, devolviendo la tabla de datos no consecutiva.
data.sample(frac = 1) 
# Con frac "fracmento" le decimos que porcentaje de la tabla queremos, con [ sample(frac = 1) ] nos devolvera el 100% de los datos.
# Si solo queremos la mitad, [ sample(frac = 0.5)]. Si queremos el 80% [ sample(frac = 0.8) ] y asi sucesivamente
data.sample(frac = 0.5)

# si queremos saber que columnas tenemos se hace con la funcion "columns"
data.columns

# Para saber los distintos tipos de variables de nuestro Data Frame, es decir nos mostrara el nombre de la columna y que tipo de datos alamacena
data.dtypes

# Si solo queremos ver los datos en un formato sin estetica, es decir sin en una tabla
data.values

# si queremos guardar cierta informacion de el DF original o otra variable. 
data2 = data.head(10)
# Se guardo las primeras 10 filas del DF orginal es data2, que solo contiene esas primeras 10 lineas. 
# Mostramos el data2
data2.head(20) # por mas que le pasemos 20, solo mostrara las 10 filas que tiene. >> ESTO CONSUME MEMORIA <<
```
